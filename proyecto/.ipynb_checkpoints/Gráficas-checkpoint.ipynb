```python
#Se importan las librerías que se van a usar
import matplotlib.pyplot as plt #Matplot sirve para graficar
import numpy as np #Numpy sirve para hacer arreglos
import seaborn as sns #Sirve para graficar densidades
plt.ion()

%matplotlib inline
```

# Variables usadas

```python
datos=np.loadtxt('datos.txt') #Se lee el archivo de lectura
x=datos[:,0] #Se guarda un arreglo de "x" de la base de datos, ya que en columna 1 está x del experimento
y=datos[:,1] #Se guarda un arreglo de "y" de la base de datos, ya que en columna 2 está y del experimento
o=datos[:,2] #Se guarda un arreglo de "barras de error" de la base de datos, ya que en columna 3 están las barras del experimento 

cadena=np.loadtxt('Res_cadena2.txt')#Se lee el archivo de resultados
like=cadena[:,0] #Se guarda un arreglo de likelihood
mn=cadena[:,1]#Se guarda un arreglo de pendiente nueva
bn=cadena[:,2]#Se guarda un arreglo de ordenada al origen nueva

print(bn)
```

# Gráfica de mnew vs bnew

```python
fig1=plt.scatter(mn, bn, color='blue') #Grafico mn contra bn en puntos
plt.xlabel('m nueva') ##Pongo los labels en x y y
plt.ylabel('b nueva') 
plt.title ('Cadena4: m vs b')#POngo el título de la figura
plt.savefig('MvsB4') #Se guarda la figura en el github
```

# Histogramas de parámetros

HISTOGRAMA DE "Mnew"

```python
plt.hist(mn, bins=60, edgecolor = 'black',  linewidth=1, color='blue') #Se grafica el histograma de la pendiente
mediam=np.mean(mn) # Se calcula la media aritmetica
print(mediam)
desvm=np.std(mn) #Se calcula la desviación estandar, que es el error asociado
print(desvm)
medianm=np.median(mn) #Se calcula la mediana de los resultados
print(medianm)
plt.title ('Histograma4: m') #Se pone el título del gráfico
plt.savefig('HistM-4') #Se guarda el gráfico
```

HISTOGRAMA DE "Mnew con densidad"

```python
#Se grafica el histograma de la pendiente con su densidad
plt.subplot(2,1,1) #En el primer gráfico se muestra la frecuencia de m
plt.hist(mn, bins=60, edgecolor = 'black',  linewidth=1, color='blue')
plt.title ('Histograma4 de m') #Se pone el título
plt.ylabel('Frecuencia') #Se ponen los títulos de los ejes

plt.subplot(2,1,2) #En el 2 gráfico se muestra la densidad de m
sns.distplot(mn, hist=True,
             bins=int(180/5), color = 'darkblue',hist_kws={'edgecolor':'black'})
plt.xlabel('Intervalos') ##Pongo los labels en x y y
plt.ylabel('Densidad') 

plt.savefig('HistM-4')#Se guarda la imagen
```

HISTOGRAMA DE "Bnew"

```python
plt.hist(bn, bins=60, edgecolor = 'black',  linewidth=1, color='blue') #Se grafica el histograma
mediab=np.mean(bn) #Se calcula la media aritmetica
print(mediab)
desvb=np.std(bn) ##Se calcula la desviación estandar, que es el error asociado
print(desvb)
medianb=np.median(bn) #Se calcula la mediana de los resultados
print(medianb)
plt.title ('Histograma4: b') #Se pone el título
plt.savefig('HistB-4') #Se guarda la imagen
```

HISTOGRAMA DE "Bnew con densidad"

```python
#Se grafica el histograma de la ordenada con su densidad
plt.subplot(2,1,1)  #En el primer gráfico se muestra la frecuencia de b
plt.hist(bn, bins=60, edgecolor = 'black',  linewidth=1, color='blue') #Se grafica el histograma
plt.title ('Histograma4 de b') #Se pone el título
plt.ylabel('Frecuencia') #Se ponen el label de y

plt.subplot(2,1,2) #En el 2 gráfico se muestra la densidad de b
sns.distplot(bn, hist=True,
             bins=int(180/5), color = 'darkblue',hist_kws={'edgecolor':'black'}) #Se grafica la densidad del histograma
plt.ylabel('Densidad') #Se pone el label de y

plt.savefig('HistB-4') #Se guarda la imagen
```

HISTOGRAMA DE "LIKELIHOOD"

```python
plt.hist(like, bins=60, edgecolor = 'black',  linewidth=1, color='blue') #Se grafica el histograma
medialike=np.mean(like) #Se calcula la media aritmetica
print(medialike)
desvlike=np.std(like) ##Se calcula la desviación estandar, que es el error asociado
print(desvlike)
medianlike=np.median(like) #Se calcula la mediana
print(medianlike)
plt.title ('Histograma4: likelihood') #Se pone el título del gráfico

plt.savefig('HistL-4') #Se guarda el gráfica
```

# Gráfica de experimento vs obtenida con b y m nuevas

```python
yn=mediam*x+mediab #Se calcula la recta con m y b finales que el programa obtinene (las cuales son las más exactas posibles)
```

```python
plt.errorbar(x, y, o, fmt='o', color='black') #Se grafican los valores del experimento

plt.plot(x,yn, color='blue',linewidth=1) #Grafico las barras de error

plt.xlabel('x') ##Pongo los labels en x y y
plt.ylabel('y') 
plt.title ('Ajuste de resultados (cadena2) del experimento') #Se pone el título del gráfico
plt.savefig('Ajuste-2') #Se guarda la gráfica
```
