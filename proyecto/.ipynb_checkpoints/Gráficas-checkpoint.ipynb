```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
plt.ion()

%matplotlib inline
```

# Variables usadas

```python
datos=np.loadtxt('datos.txt')
x=datos[:,0] 
y=datos[:,1]
o=datos[:,2]

cadena=np.loadtxt('Res_cadena2.txt')
like=cadena[:,0] 
mn=cadena[:,1]
bn=cadena[:,2]

print(bn)
```

# Gráfica de mnew vs bnew

```python
fig1=plt.scatter(mn, bn, color='blue')
plt.xlabel('m nueva') ##Pongo los labels en x y y
plt.ylabel('b nueva') 
plt.title ('Cadena4: m vs b')
plt.savefig('MvsB4')
```

# Histogramas de parámetros

```python
plt.hist(mn, bins=60, edgecolor = 'black',  linewidth=1, color='blue')
mediam=np.mean(mn) #Media aritmetica
print(mediam)
desvm=np.std(mn) #DEsv estandar, que es el error asociado
print(desvm)
medianm=np.median(mn) #DEsv estandar, que es el error asociado
print(medianm)
plt.title ('Histograma4: m')
plt.savefig('HistM-4')
```

```python
plt.subplot(2,1,1)
plt.hist(mn, bins=60, edgecolor = 'black',  linewidth=1, color='orange')
plt.title ('Histograma4 de m')
plt.ylabel('Frecuencia') 

plt.subplot(2,1,2)
sns.distplot(mn, hist=True,
             bins=int(180/5), color = 'darkblue',hist_kws={'edgecolor':'black'})
plt.xlabel('Intervalos') ##Pongo los labels en x y y
plt.ylabel('Densidad') 

plt.savefig('HistM-4')
```

```python
plt.hist(bn, bins=60, edgecolor = 'black',  linewidth=1, color='blue')
mediab=np.mean(bn) #Media aritmetica
print(mediab)
desvb=np.std(bn) #DEsv estandar, que es el error asociado
print(desvb)
medianb=np.median(bn) #DEsv estandar, que es el error asociado
print(medianb)
plt.title ('Histograma4: b')
plt.savefig('HistB-4')
```

```python
plt.subplot(2,1,1)
plt.hist(bn, bins=60, edgecolor = 'black',  linewidth=1, color='orange')
plt.title ('Histograma4 de b')
plt.ylabel('Frecuencia') 

plt.subplot(2,1,2)
sns.distplot(bn, hist=True,
             bins=int(180/5), color = 'darkblue',hist_kws={'edgecolor':'black'})
plt.ylabel('Densidad') 

plt.savefig('HistB-4')
```

```python
plt.hist(like, bins=60, edgecolor = 'black',  linewidth=1, color='blue')
medialike=np.mean(like) #Media aritmetica
print(medialike)
desvlike=np.std(like) #DEsv estandar, que es el error asociado
print(desvlike)
medianlike=np.median(like) #DEsv estandar, que es el error asociado
print(medianlike)
plt.title ('Histograma4: likelihood')

plt.savefig('HistL-4')
```
