```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline

```

```python
datos=np.loadtxt('datos.txt')
x=datos[:,0] 
y=datos[:,1]
o=datos[:,2]

cadena=np.loadtxt('Resultados.txt')
like=cadena[:,0] 
mn=cadena[:,1]
bn=cadena[:,2]

print(bn)
```

```python
plt.scatter(mn, bn, color='violet')
plt.xlabel('m nueva') ##Pongo los labels en x y y
plt.ylabel('b nueva') 
plt.title ('Resultados m vs b')
```

```python


plt.hist(mn, bins=60, edgecolor = 'black',  linewidth=1, color='red')
mediam=np.mean(mn) #Media aritmetica
print(mediam)
desvm=np.std(mn) #DEsv estandar, que es el error asociado
print(desvm)
medianm=np.median(mn) #DEsv estandar, que es el error asociado
print(medianm)

```

```python
plt.hist(bn, bins=60, edgecolor = 'black',  linewidth=1, color='yellow')
mediab=np.mean(bn) #Media aritmetica
print(mediab)
desvb=np.std(bn) #DEsv estandar, que es el error asociado
print(desvb)
medianb=np.median(bn) #DEsv estandar, que es el error asociado
print(medianb)
```

```python
plt.hist(like, bins=60, edgecolor = 'black',  linewidth=1, color='orange')
medialike=np.mean(like) #Media aritmetica
print(medialike)
desvlike=np.std(like) #DEsv estandar, que es el error asociado
print(desvlike)
medianlike=np.median(like) #DEsv estandar, que es el error asociado
print(medianlike)
```

```python
yn=mediam*x+mediab
```

```python
plt.errorbar(x, y, o, fmt='o')

plt.plot(x,yn, color='black') #Grafico las barras de error

plt.xlabel('x') ##Pongo los labels en x y y
plt.ylabel('y') 
plt.title ('Resultados del experimento')
```
