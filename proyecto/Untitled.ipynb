```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
plt.ion()

%matplotlib inline
```

```python
datos=np.loadtxt('datos.txt')
x=datos[:,0] 
y=datos[:,1]
o=datos[:,2]

cadena=np.loadtxt('Res_cadena2.txt')
like=cadena[:,0] 
mn=cadena[:,1]
bn=cadena[:,2]

print(bn)
```

# Gráfica de mnew vs bnew

```python
fig1=plt.scatter(mn, bn, color='green')
plt.xlabel('m nueva') ##Pongo los labels en x y y
plt.ylabel('b nueva') 
plt.title ('Cadena2: m vs b')
plt.savefig('MvsB2')
```

# Histogramas de parámetros

```python
plt.hist(mn, bins=60, edgecolor = 'black',  linewidth=1, color='green')
mediam=np.mean(mn) #Media aritmetica
print(mediam)
desvm=np.std(mn) #DEsv estandar, que es el error asociado
print(desvm)
medianm=np.median(mn) #DEsv estandar, que es el error asociado
print(medianm)
plt.title ('Histograma2: m')
plt.savefig('HistM-2')
```

```python
plt.hist(bn, bins=60, edgecolor = 'black',  linewidth=1, color='green')
mediab=np.mean(bn) #Media aritmetica
print(mediab)
desvb=np.std(bn) #DEsv estandar, que es el error asociado
print(desvb)
medianb=np.median(bn) #DEsv estandar, que es el error asociado
print(medianb)
plt.title ('Histograma2: b')
plt.savefig('HistB-2')
```

```python
plt.hist(like, bins=60, edgecolor = 'black',  linewidth=1, color='green')
medialike=np.mean(like) #Media aritmetica
print(medialike)
desvlike=np.std(like) #DEsv estandar, que es el error asociado
print(desvlike)
medianlike=np.median(like) #DEsv estandar, que es el error asociado
print(medianlike)
plt.title ('Histograma2: likelihood')

plt.savefig('HistL-2')
```

# Gráfica de experimento vs obtenida con b y m nuevas

```python
yn=mediam*x+mediab
```

```python
plt.errorbar(x, y, o, fmt='o', color='green')

plt.plot(x,yn, color='black') #Grafico las barras de error

plt.xlabel('x') ##Pongo los labels en x y y
plt.ylabel('y') 
plt.title ('Ajuste de resultados del experimento')
plt.savefig('Ajuste-2')
```
